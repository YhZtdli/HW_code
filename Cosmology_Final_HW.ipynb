```python
import numpy as np
import scipy.special as sp
import matplotlib.pyplot as plt
```

```python
def C_func(r):
    return r**2

def A_func(r):
    return 1-2/r

def u_func(P):
    return np.sqrt(C_func(P)/A_func(P))

def Q_func(P):
    return np.sqrt((P-2)*(P+6))

def alpha_fun(P):
    phi1 = np.arcsin(np.sqrt(2*Q_func(P)/(3*P-6+Q_func(P))))
    m = (Q_func(P)-P+6)/(2*Q_func(P))
    return -np.pi+4*np.sqrt(P/Q_func(P))*sp.ellipkinc(phi1, m)
```

```python
P_list = np.linspace(3.1, 20, 1000)
```

```python
plt.plot(u_func(P_list), alpha_fun(P_list))
plt.show()
```

```python
ubar = 3*np.sqrt(3)
def alpha_SDL(u):
    return -np.log(u/ubar-1)+np.log(216*(7-4*np.sqrt(3)))-np.pi
```

```python
u_list = np.linspace(ubar+0.01, 20, 1000)
plt.axvline(x=ubar, linestyle = ':', color='red', lw=0.5)
plt.plot(u_list, alpha_SDL(u_list))
plt.show()
```

```python
plt.plot(u_func(P_list), alpha_fun(P_list), lw = 0.9, label=r'$\alpha_{\rm real}$')
plt.axvline(x=ubar, linestyle = ':', color='red', lw=0.5)
plt.plot(u_list, alpha_SDL(u_list), lw = 0.9, linestyle = '-.', label=r'$\alpha_{\rm SDL}$')
plt.plot(u_list, 4/u_list, lw = 0.9, linestyle = '--', label=r'$\alpha_{\rm WDL}$')
plt.xlim(5, 20)
plt.ylim(0, 6.2)
plt.xlabel(r'$\sqrt{C(P)/A(P)}\ [M]$')
plt.ylabel(r'$\alpha$')
plt.legend()
plt.show()
```
